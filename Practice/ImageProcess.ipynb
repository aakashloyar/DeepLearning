```python
from PIL import Image
import matplotlib.pyplot as plt

# Open the image and convert to grayscale
image = Image.open('myimg.png').convert('L')

# Display the grayscale image using matplotlib
plt.imshow(image, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```

Now we will see how to convert image to to it's pixel values

```python
pixel_values=list(image.getdata());
```

Now plotting a histogram on the image pixel data

```python
plt.hist(pixel_values, bins=256, range=(0, 256), density=True, color='gray')
plt.title('Grayscale Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
#here we are taking the pixel value of image at given x and y value

x, y = 100, 100
pixel = image.getpixel((x, y))
print(f"Pixel value at ({x},{y}): {pixel}")
```

```python
# Get the image size (width, height)
width, height = image.size

# Calculate total pixels
total_pixels = width * height
print(f"Image Resolution: {width} x {height}")
print(f"Total Pixels: {total_pixels}")
```
